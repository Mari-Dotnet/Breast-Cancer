```python
def cor_relation(data,threshold):
    corr_olumn=[]
    corr_matrix=data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                   colname=corr_matrix.columns[i]
                   corr_olumn.append(colname)
    return corr_olumn
            
```

```python
corr_leation_column=cor_relation(df,0.85)
```

```python
column_series=pd.Series(corr_leation_column)
```

```python
Remove_corr_column_list=column_series.unique()
```

```python
Remove_corr_column_list
```

```python
df=df.drop(labels=Remove_corr_column_list,axis=1)
```
